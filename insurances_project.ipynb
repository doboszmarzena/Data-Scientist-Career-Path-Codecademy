```python
#1. I found out the average age of the patients in the dataset.
aver_insurance = []
with open('insurance.csv') as insurance:
    insurance_file = csv.DictReader(insurance)
    for i in insurance_file:
        aver_insurance.append(i['age'])
```

```python
insurance_int = [int(i) for i in aver_insurance]
insurance_int = sum(insurance_int) // len(aver_insurance)
print(f"The average age of the patients in the datase is: {insurance_int} years")
```

```python
import csv
#3. I looked at the different costs between smokers vs. non-smokers.
smokers = []
non_smokers = []
with open('insurance.csv') as insurance:
    insurance_file = csv.DictReader(insurance)
    for i in insurance_file:
        if "yes" in i['smoker']:
            smokers.append(i['charges'])
        elif "no" in i['smoker']:
            non_smokers.append(i['charges'])
```

```python
smokers_charges = [float(i) for i in smokers]
smokers_charges = sum(smokers_charges)
non_smokers_charges = [float(i) for i in non_smokers]
non_smokers_charges = sum(non_smokers_charges)
if smokers_charges >= non_smokers_charges:
    between_charges = smokers_charges - non_smokers_charges
    print(f"Different costs between smokers vs. non-smokers : {between_charges}")
else:
    between_charges = non_smokers_charges - smokers_charges
    print(f"Different costs between non-smokers vs. smokers : {round(between_charges, 2)}")
    
```
